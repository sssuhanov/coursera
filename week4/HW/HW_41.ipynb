#### Подключим все необходимые библиотеки

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
import random
%matplotlib inline
```

```python
b = 0.774972521011
```

```python
# возьмем выборку из 1000 значений
r = sts.rice.rvs(b, size=1000)
```

```python
# Рисуем гастраграмму
plt.hist(r, bins=30, density=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')

# на том же графике построим теоретическую плотность распределения:
x = np.linspace(0,5,1000)
pdf = sts.rice.pdf(x, b)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
# зададим функцию генерирующую 1000 выборочных средних взятых из выборки размером n
# помещаем выборочные средние в вектор
def random_mean_rice(n):
    new_sample = np.zeros(1000)
    for x in np.arange(0, 1000):
        r = sts.rice.rvs(b, size=n)
        new_sample[x] = np.mean(r)
    return new_sample
```

```python
# рисуем график
def hist_start(n):
    # генерируем вектор с 1000 значений выбороных средних
    new_sample = random_mean_rice(n)
    
    # рисуем гистограмму
    bins = np.linspace(0, 3, 100)    
    plt.hist(new_sample, bins, density=True)
    plt.ylabel('fraction of samples')
    plt.xlabel('$x$')
    
    # определяем среднее и стандартное отклонение вектора
    mean = np.mean(new_sample)
    std = np.std(new_sample)
    print("Среднее: ", mean)
    print("Стандартное отклонение: ", std)

    # строим график плотности нормального распределения:
    left = sts.rice.ppf(0.01, b) 
    right = sts.rice.ppf(0.99, b)
    x = np.linspace(left,  right, 100)
    pdf = sts.norm.pdf(x, mean, std)
    plt.plot(x, pdf, 'r-', label='normal pdf', alpha=0.9)
    plt.legend()
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
```

```python
hist_start(5)
```

```python
hist_start(10)
```

```python
hist_start(50)
```

### Как видно из графиков выше чем больше объем выборок для выборочных средних, тем сильнее гистограмма выборочных средних напоминает нормальное расрпеделение.
